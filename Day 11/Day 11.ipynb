# Topic- Corelation between attrition and other variables

In order to find the interdependency of the variables DistanceFromHome,Education ,Gender,NumCompaniesWorked,PercentSalaryHike ,MonthlyIncome,YearsSinceLastPromotion,TotalWorkingYears,TrainingTimesLastYear, YearsAtCompany, YearsWithCurrManager ,Age from that of Attrition, we executed the Correlation Analysis as follows :

```python
import pandas as pd
import numpy as np
dataset=pd.read_csv("general_data.csv")
dataset.head()
```

```python
from sklearn import preprocessing
Label_encoder=preprocessing.LabelBinarizer()
dataset["Attrition"]=Label_encoder.fit_transform(dataset["Attrition"])
from scipy.stats import pearsonr
stats,p=pearsonr(dataset.Attrition,dataset.Age)
print(stats,p)
```

```python
stats,p=pearsonr(dataset.Attrition,dataset.DistanceFromHome)
print(stats,p)
```

```python
stats,p=pearsonr(dataset.Attrition,dataset.Education)
print(stats,p)
```

```python
from sklearn import preprocessing
Label_encoder1=preprocessing.LabelBinarizer()
dataset["Gender"]=Label_encoder1.fit_transform(dataset["Gender"])
stats,p=pearsonr(dataset.Attrition,dataset.Gender)
print(stats,p)
```

```python
stats,p=pearsonr(dataset.Attrition,dataset.PercentSalaryHike)
print(stats,p)
```

```python
stats,p=pearsonr(dataset.Attrition,dataset.StockOptionLevel)
print(stats,p)
```

```python
stats,p=pearsonr(dataset.Attrition,dataset.TrainingTimesLastYear)
print(stats,p)
```

```python
stats,p=pearsonr(dataset.Attrition,dataset.YearsAtCompany)
print(stats,p)
```

```python
stats,p=pearsonr(dataset.Attrition,dataset.YearsSinceLastPromotion)
print(stats,p)
```

```python
stats,p=pearsonr(dataset.Attrition,dataset.YearsWithCurrManager)
print(stats,p)
```
