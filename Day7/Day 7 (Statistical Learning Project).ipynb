<font color=red>__Step 5 â€“ Statistical Tests (Mann-Whitney)__</font>

```python
Attrition_Yes=dataset[dataset["Attrition"]=='Yes']
Attrition_Yes.head()
```

```python
Attrition_Yes.columns
```

```python
Attrition_No=dataset[dataset["Attrition"]=='No']
Attrition_No.head()
```

```python
Attrition_No.columns
```

__Attrition vs Distance From Home__

```python
from scipy.stats import mannwhitneyu 
stat, p=mannwhitneyu(Attrition_Yes.DistanceFromHome,Attrition_No.DistanceFromHome)
print(stat, p)
```

<font color="Blue">As the P value is > 0.05, the H0 is accepted and Ha is rejected.
    
H0: There is no significant differences in the Distance From Home between attrition (Y) and attirition
(N)
    
Ha: There is significant differences in the Distance From Home between attrition (Y) and attirition (N)</font>

<font color=red>__Step 7 â€“ Unsupervised Learning - Correlation Analysis__</font>


In order to find the interdependency of the variables DistanceFromHome, MonthlyIncome, Age ,TotalWorkingYears, YearsAtCompany, Years Since Last Promotion, YearsWithCurrManager from that of Attrition, we executed the Correlation Analysis as follows: 

```python
from scipy.stats import pearsonr
dataset['TotalWorkingYears']=dataset['TotalWorkingYears'].fillna(11.28)
```

__Attrition & Distance From Home__

```python
from sklearn import preprocessing
Label_encoder=preprocessing.LabelBinarizer()
dataset["Attrition"]=Label_encoder.fit_transform(dataset["Attrition"])
stats, p=pearsonr(dataset.Attrition, dataset.DistanceFromHome)
print(stats, p)
```

<font color=green>___H0-There is no significant corelation between Attrition and Distance from home.___
    
___Ha-There is significant corelation between Attrition and Distance from home.___
</font>

As the P value of 0.518 is > 0.05, we are accepting H0 and hence thereâ€™s no significant correlation
between Attrition & DistanceFromHome.
